```python
import matplotlib.pyplot as plt
from scipy.io import loadmat
import pandas as pd
from scipy.stats import pearsonr
```

```python
ADNI_conn = loadmat('../data/dk_ADNI_averaged.mat')['connectome_avg']
meanACS_conn = loadmat('../data/dk_conn_meanACS69.mat')['meanACS']
mean80_conn = pd.read_csv('../data/mean80_fibercount.csv')
mean80_conn = mean80_conn.set_index(mean80_conn.columns)
```

```python
import importlib.util

raj_utils_path = "../utils/atlas_utils.py"
spec = importlib.util.spec_from_file_location("atlas-utils.load_atlas", raj_utils_path)
raj_utils = importlib.util.module_from_spec(spec)
spec.loader.exec_module(raj_utils)
```

```python
# load desikan utils info. Index is in Free-Surfer ordering
dk_atlas_info = raj_utils.load_atlas(atlas="DK").set_index('Name')

# Get the cortical regions
right_cortical = dk_atlas_info.loc[dk_atlas_info['Hemisphere'] == 'Right'].loc[dk_atlas_info['Cortex'] == 'cortical']
left_cortical = dk_atlas_info.loc[dk_atlas_info['Hemisphere'] == 'Left'].loc[dk_atlas_info['Cortex'] == 'cortical']

# Get the subcortical regions
left_subcortical = dk_atlas_info.loc[dk_atlas_info['Hemisphere'] == 'Left'].loc[dk_atlas_info['Cortex'] == 'subcortical']
right_subcortical = dk_atlas_info.loc[dk_atlas_info['Hemisphere'] == 'Right'].loc[dk_atlas_info['Cortex'] == 'subcortical']

#Make DFs following the order above
LRLR = pd.concat([left_cortical, right_cortical, left_subcortical, right_subcortical])
LRRL = pd.concat([left_cortical, right_cortical, right_subcortical, left_subcortical])

#Get the region names of these dataframes
LRLR_names = list(LRLR.index)
LRRL_names = list(LRRL.index)
```

## Creating dataframes for connectivities

```python
plt.figure(figsize=(20,20))

plt.subplot(1, 3, 1)
plt.title('ADNI averaged connectome')
_ = plt.matshow(ADNI_conn, fignum=False)

plt.subplot(1, 3, 2)
plt.title("mean ACS69")
_ = plt.matshow(meanACS_conn, fignum=False)

plt.subplot(1, 3, 3)
plt.title("mean80 fibercount")
_ = plt.matshow(mean80_conn, fignum=False)
```

From the above exercise:
- `ADNI` is LRLR, CCSS
- `meanACS69` is LRRL, CCSS
- `mean80` is LRLR, SSCC

Where L/R mean Left/Right hemispheres. C/S means Cortical/Subcortical structures.

`ADNI` and `meanACS69` were commonly used in the lab and it is expected (although not 100% certain) that the order follows FreeSurfer (therefore not alphabetical).

`mean80` is already in the dictionary form although right now I'm not able to track it down its origin."

### 1. Fixing ADNI ordering to FreeSurfer

```python
ADNI_conn_df = pd.DataFrame(ADNI_conn, columns=LRLR_names, index=LRLR_names)
ADNI_conn_df.head(3)
```

```python
# ADNI connectivity is already in FreeSurfer order, but for completion:
ADNI_conn_fs_order = ADNI_conn_df[LRLR_names].loc[LRLR_names]
```

### 2. Fixing meanACS ordering to FreeSurfer

```python
meanACS_conn_df = pd.DataFrame(meanACS_conn, columns=LRRL_names, index=LRRL_names)
meanACS_conn_df.head(3)
```

```python
meanACS_conn_fs_order = meanACS_conn_df[LRLR_names].loc[LRLR_names]
meanACS_conn_fs_order.head(3)
```

### 3. Fixing mean80 ordering to FreeSurfer

```python
#mean80 is already a pandas dataframe but 2 names are incomplete / different
mean80_conn_df = pd.DataFrame(mean80_conn)

mean80_conn_df = mean80_conn_df.rename(columns={'Left-VentralDC':'Left-VentralDC/Hypothalamus', 'Right-VentralDC':'Right-VentralDC/Hypothalamus'})
mean80_conn_df = mean80_conn_df.rename(index={'Left-VentralDC':'Left-VentralDC/Hypothalamus', 'Right-VentralDC':'Right-VentralDC/Hypothalamus'})

mean80_conn_df.head(3)
```

```python
mean80_conn_fs_order = mean80_conn_df[LRLR_names].loc[LRLR_names]
mean80_conn_fs_order.head(3)
```

## 4. New plots

```python
plt.figure(figsize=(20,20))

plt.subplot(1, 3, 1)
plt.title('Reordered ADNI averaged connectome')
_ = plt.matshow(ADNI_conn_fs_order, fignum=False)

plt.subplot(1, 3, 2)
plt.title("Reordered mean ACS69")
_ = plt.matshow(meanACS_conn_fs_order, fignum=False)

plt.subplot(1, 3, 3)
plt.title("Reordered mean80 fibercount")
_ = plt.matshow(mean80_conn_fs_order, fignum=False)
```

## 5. Export to CSV and pickles

```python
ADNI_conn_fs_order.to_csv('../data/postprocessing/ADNI_conn_fs_order.csv')
meanACS_conn_fs_order.to_csv('../data/postprocessing/meanACS_conn_fs_order.csv')
mean80_conn_fs_order.to_csv('../data/postprocessing/mean80_conn_fs_order.csv')
```
